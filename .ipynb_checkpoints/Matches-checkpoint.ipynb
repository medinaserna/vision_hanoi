```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap='gray'):
    fig = plt.figure(figsize = (12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
hanoi = cv2.imread('DATA/hanoi6.jpg', 0)
escritorio = cv2.imread('DATA/hanoi7.jpg',0)
```

```python
#display(escritorio)
```

```python
# aqui creo un objeto que va a extraer los parametros
orb = cv2.ORB_create()
```

```python
key1, des1 = orb.detectAndCompute(hanoi, None)
key2, des2 = orb.detectAndCompute(escritorio, None)
```

```python
#este metodo es el brute force, es el mas basico y el menos acertado
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)
```

```python
matches = bf.match(des1, des2)
```

```python
len(matches)
```

```python
hanoi_matches = cv2.drawMatches(hanoi, key1, escritorio, key2, matches, None, flags = 2)
```

```python
display(hanoi_matches)
```

```python
sift =  cv2.xfeatures2d.SIFT_create()
```

```python
kp1, des1 = sift.detectAndCompute(hanoi, None)
kp2, des2 = sift.detectAndCompute(escritorio, None)
```

```python
bf = cv2.BFMatcher()
```

```python
#este metodo el knn es mucho mejor en el tutorial, pero en mi caso no pega
matches = bf.knnMatch(des1, des2, k=2)
```

```python
good = []
for match1, match2 in matches:
#IF THE DISTANCE OF MATCH1 IS LESS THAN 75% OF MATCH2 IS A GOOD DESCRIPTOR
    if match1.distance < 0.75*match2.distance:
        good.append([match1])
```

```python
sift_matches = cv2.drawMatchesKnn(hanoi, kp1, escritorio, kp2, good, None, flags = 2)
```

```python
display (sift_matches)
```

```python
#FLANN
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict (checks = 50)
```

```python
flann = cv2.FlannBasedMatcher(index_params, search_params)
```

```python
matches = flann.knnMatch(des1, des2, k=2)
```

```python
matchesMask = [[0,0] for i in range(len(matches))]
```

```python
#good = []
for i, (match1, match2) in enumerate (matches):
#IF THE DISTANCE OF MATCH1 IS LESS THAN 70% OF MATCH2 IS A GOOD DESCRIPTOR
    if match1.distance < 0.7*match2.distance:
        matchesMask[i] = [1,0]
        
        #good.append([match1])
```

```python
draw_params = dict(matchColor =(0,255,0),
                  singlePointColor = (255,0,0),
                  matchesMask = matchesMask,
                  flags = 2)
```

```python
flann_matches = cv2.drawMatchesKnn(hanoi, kp1, escritorio, kp2,matches, None, **draw_params)
display(flann_matches)
```
