```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap='gray'):
    fig = plt.figure(figsize = (10,12))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
hanoi = cv2.imread('DATA/ROI_selected.jpg')
```

```python
blur_hanoi = cv2.medianBlur(hanoi,1)
blur_hanoi = cv2.cvtColor(blur_hanoi, cv2.COLOR_BGR2RGB)
```

```python
gray_hanoi = cv2.cvtColor(blur_hanoi, cv2.COLOR_BGR2GRAY)
```

```python
display(blur_hanoi)
```

```python
display(gray_hanoi)
```

```python
ret, hanoi_thres = cv2.threshold(gray_hanoi, 40, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
display(hanoi_thres)
```
